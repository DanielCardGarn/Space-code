```python
from math import radians, cos, sin, asin, sqrt

# Función para calcular la distancia entre dos coordenadas geográficas
def haversine(lat1, lon1, lat2, lon2):
    # Convertir coordenadas de grados a radianes
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Fórmula de Haversine
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371  # Radio de la Tierra en km
    return c * r
```

```python
# Supongamos que tomamos el primer sitio como referencia
base_site = launch_sites_df.iloc[0]

# Crear una nueva columna con la distancia al sitio base
launch_sites_df['DistanceFromBase_km'] = launch_sites_df.apply(
    lambda row: haversine(base_site['Lat'], base_site['Long'], row['Lat'], row['Long']),
    axis=1
)

launch_sites_df[['Launch Site', 'DistanceFromBase_km']]
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
